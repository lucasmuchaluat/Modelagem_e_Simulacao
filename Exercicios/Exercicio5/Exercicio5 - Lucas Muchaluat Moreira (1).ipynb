**Exercício 5 - Derivadas, Equações Diferenciais e o Método de Euler**
   
Neste exercício, vamos entender melhor os conceitos de derivada e de equações diferenciais. Primeiramente, você já deve estar se acostumando com o conceito de derivada, que chamamos de "taxa de variação instantânea". Dados os exercícios feitos até aqui, você já deve ter entendido que a taxa de variação temporal de uma função no intervalo de tempo $t_1$ e $t_2$ é dada por:
   
$\frac{\Delta f}{\Delta t}=\frac{f(t_2)-f(t_1)}{t_2-t_1}$
   
Pois bem, já está ficando bem chato repetir que você ainda vai aprender em MatVar que quando $t_2$ é muito próximo de $t_1$ a taxa de variação será chamada de derivada (já sacou, né?). Mais uma vez:
   
$\frac{df}{dt}=\lim_{t_2 \to t_1}\frac{f(t_2)-f(t_1)}{t_2-t_1}$
   
Você deve estar muito ansioso para aprender como resolver a equação acima algebricamente para vários tipos de funcões, mas, por enquanto, vamos trabalhar o conceito da taxa de variação em um ponto, e construir uma série temporal de taxas para vários pontos. Ou seja, se começarmos a calcular as taxas de variação de uma determinada função quanto $t=0$ e formos incrementando esse $t$ a passos bem pequenos, poderemos calcular a taxa de variação a cada intervalo $\Delta t$, construindo uma lista de taxas de variação. A essa lista de taxas de variação, daremos o nome de "função derivada".


**Item 1 - Implementando numericamente a função derivada**

Vamos começar implementando uma função bem simples e muito conhecida por você: $f(t)=t^2$

Na célula abaixo, escreva uma função em Python para a implementação dessa complicada função.

```python
#implemente seu código aqui
def funcao(t):
    return t**2
```

1a) Para se certificar de que sua função funciona, plote um gráfico de $f(t)$ para $t$ variando de 0 a 10 com intervalos de tempo de 0,01 $s$. Use a função "arange" do módulo "numpy" para criar uma lista de tempo.

```python
#implemente seu código aqui
import numpy as np
import matplotlib.pyplot as plt

tempo=np.arange(0,10,0.01)

x=[]
for i in tempo:
    x.append(funcao(i))

plt.figure(figsize = [12, 4])
plt.title('f=t**2')
plt.plot(tempo,x)
plt.xlabel("t")
plt.ylabel("f(t)")
plt.grid(True)
plt.show()
```

1b) Agora que sua função já está funcionando, vamos implementar o cálculo das taxas de variação de $f(t)$ para um determinado intervalo de tempo. Para isso, vamos criar uma função chamada TaxadeVariacao(t, delta_t), que deve retornar a taxa de variação no instante t considerando um determinado $\Delta t$. Para isso, use a função y(t) que você já implementou!

```python
#implemente seu código aqui
def TaxadeVariacao(t,delta_t):
    return (funcao(t+delta_t)-funcao(t))/delta_t
```

```python
#implemente seu código aqui
import numpy as np
y=[]
tx_var=[]
contador=np.arange(0,10,0.01)

for e in contador:
    y.append(funcao(e))
    tx_var.append(TaxadeVariacao(e,0.01))

plt.figure(figsize = [12, 4])
plt.title('Derivada f=t**2')
plt.plot(contador,tx_var)
plt.ylabel("Taxa de Variação")
plt.xlabel("t")
plt.grid(True)
plt.show()
```

**Item 2 - Resolvendo mais uma equação diferencial pelo método de Euler, mas com uma taxa de variação que também é função do tempo!**

Vamos imaginar que queiramos saber qual a solução de uma equação diferencial do tipo $\frac{dy}{dt}=f(y, t)$.

Por exemplo, uma equação que seja dada pela expressão abaixo:

$\frac{dy}{dt}=a\cdot y\cdot[1+2\cdot sin(\omega \cdot t)]$

Essa equação representa a evolução diária da população de bactérias em um determinado meio, que depende da quantidade de bactéricas existentes e de uma função senoidal que representa a sazonalidade do processo.

Pois bem, vamos encontrar qual função $y(t)$ é solução da equação diferencial acima. Para isso, vamos usar, mais uma vez, o método de Euler.

Primeiramente, implemente abaixo uma função chamada "EquacaoDiferencial(y, t, a, w)", que recebe os quatro parâmetros indicados nessa ordem. Essa função deve retornar o lado direito da equação $dy/dt$.

```python
#implemente seu código aqui
import math
def EquacaoDiferencial(y,t,a,w):
    return a*y*(1+2*math.sin(w*t))
```

Agora que sua função "EquacaoDiferencial" está implementada, vamos calcular a série temporal da função y(t) que satisfaz a equação diferencial que estamos tentando resolver. Os valores da função y(t) estarão numa lista, com cada elemento correspondendo a cada instante de tempo considerado. Abaixo, implemente o código do método de Euler (o mesmo que você usou no Projeto 1) executando a função "EquacaoDiferencial" e usando os seguintes parâmetros:

$y(0) = 5$ (Esta é a população inicial de bactérias)

$a = 0,007$

$\omega=0,05$

$\Delta t=0,001$ dias (nosso $\Delta t$ é um milésimo de um 1 dia)

Para sua implementação, use 400 dias de simulação.

Qual a população de bactérias após 1 ano (365 dias)?


```python
#implemente seu código aqui
import numpy as np
import matplotlib.pyplot as plt

delta_t=0.001
lista_tempo=np.arange(0,400,0.001)
Y=[5]
for j in range(1,len(lista_tempo)):
    Y.append(Y[j-1]+EquacaoDiferencial(Y[j-1],0.001,0.007,0.05)*0.001)

plt.figure(figsize = [14, 6])
plt.plot(lista_tempo, Y)
plt.xlabel("tempo (dias)")
plt.ylabel("Quantidade de bactérias")
plt.title('Solução da Equação Diferencial (Método de Euler)')
plt.grid(True)
plt.show()
```
